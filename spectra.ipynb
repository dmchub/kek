```python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

```python
f = open("OUTCAR")
data = []
for line in f:
    if "dielectric  constant" in line:
        clean = line.strip().split()
        tok = [x for x in clean if x]
        numbers = [float(i) for i in tok[:3]]
        data.append(numbers)
tutu = np.array(data)
plt.figure(figsize=(13,10))
plt.plot(tutu[:15,0], tutu[:15,1], label='Действительная часть')
plt.plot(tutu[:15,0], tutu[:15,2], label='Мнимая часть')
plt.xlabel('Энергия, эВ')
plt.ylabel('Диэлектрическая проницаемость')
plt.legend(loc=5, prop={'size': 23})
plt.grid(True)
plt.show()
```
